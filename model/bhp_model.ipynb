###  Hàm lọc outliers theo `price_per_sqft`

Hàm `remove_pps_outliers(df)` có nhiệm vụ loại bỏ các outlier (giá bất thường) trong mỗi khu vực (`location`) dựa trên phân phối chuẩn của giá trên mỗi mét vuông (`price_per_sqft`).

- Trong từng khu vực, tính **giá trị trung bình (mean)** và **độ lệch chuẩn (standard deviation)**.
- Chỉ giữ lại các điểm dữ liệu có `price_per_sqft` nằm trong khoảng `[mean - std, mean + std]`.
- Các điểm nằm ngoài khoảng này được coi là outliers và bị loại bỏ.

Hàm trả về một DataFrame mới đã loại bỏ outliers, giúp mô hình học hiệu quả và chính xác hơn.

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
df7 = remove_pps_outliers(df6)
df7.shape
```

###  Hàm trực quan hoá mối quan hệ giữa diện tích và giá bán theo BHK

Hàm `plot_scatter_chart(df, location)` có chức năng hiển thị biểu đồ phân tán (scatter plot) giữa **diện tích tổng (`total_sqft`)** và **giá bán (`price`)** cho các căn hộ 2 BHK và 3 BHK tại một khu vực cụ thể (`location`).

- Màu **xanh dương** đại diện cho các căn hộ **2 BHK**.
- Màu **xanh lá** và dấu **+** đại diện cho các căn hộ **3 BHK**.
- Trục X: diện tích tổng (m²).
- Trục Y: giá bán.

Biểu đồ giúp nhận diện trực quan sự khác biệt về giá giữa các loại BHK trong cùng một khu vực và phát hiện các bất thường nếu có.


```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) & (df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    matplotlib.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price,color='blue', label='2 BHK', s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price,marker='+', color='green', label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price")
    plt.title(location)
    plt.legend()
    plt.show()

plot_scatter_chart(df7, "Hebbal")
```

### Loại bỏ các bất thường về giá theo số phòng ngủ (BHK)

Hàm `remove_bhk_outliers(df)` được sử dụng để loại bỏ các bất thường trong dữ liệu giá trên mỗi mét vuông (`price_per_sqft`) dựa trên số lượng phòng ngủ (`bhk`) tại từng khu vực (`location`).  
Ý tưởng chính là:

- Với mỗi khu vực (`location`), ta nhóm dữ liệu theo số phòng ngủ (`bhk`).
- Ta tính:
  - Trung bình (`mean`)
  - Độ lệch chuẩn (`std`)
  - Số lượng dòng (`count`)  
  của `price_per_sqft` cho mỗi nhóm `bhk`.
- Với mỗi nhóm `bhk`, nếu có dữ liệu của nhóm `bhk - 1` và số lượng của nhóm đó > 5:
  - Các căn có cùng `bhk` nhưng giá trên mỗi mét vuông thấp hơn **trung bình của nhóm nhỏ hơn (`bhk - 1`)** sẽ bị xem là bất thường và loại bỏ.
- Cuối cùng, hàm trả về DataFrame mới với các dòng bị loại bỏ.

 Hàm này giúp đảm bảo tính hợp lý trong dữ liệu:  
Ví dụ, căn hộ 3 phòng ngủ không nên có giá thấp hơn căn hộ 2 phòng ngủ trong cùng khu vực.

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            } 
        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

df8 = remove_bhk_outliers(df7)
df8.shape
```

### Model Building

```python
X = df12.drop('price', axis='columns')
X.head()
```

```python
y = df12.price
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr_clf = LinearRegression()
lr_clf.fit(X_train,y_train)
lr_clf.score(X_test,y_test)
```

**ShuffleSplit: là kỹ thuật chia dữ liệu thành nhiều lần train/test khác nhau bằng cách xáo trộn dữ liệu ngẫu nhiên trước mỗi lần chia**

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
cross_val_score(LinearRegression(), X, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

from sklearn.pipeline import Pipeline
def find_best_model_using_gridsearchcv_old(X, y):
    algos = {
        'linear_regression' : {
            'model': LinearRegression(),
            'params': {
                
            }
        },
        'lasso' : {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree' : {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['squared_error', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'], config['params'], cv=cv, return_train_score=False)
        gs.fit(X,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })
    return pd.DataFrame(scores, columns=['model','best_score', 'best_params'])
find_best_model_using_gridsearchcv_old(X, y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

def find_best_model_using_gridsearchcv(X, y):
    models = [
        ('linear_regression', Pipeline([
            ('scaler', StandardScaler()),
            ('model', LinearRegression())
        ]), {
            # LinearRegression không cần param nào
        }),

        ('lasso', Pipeline([
            ('scaler', StandardScaler()),
            ('model', Lasso())
        ]), {
            'model__alpha': [1, 2],
            'model__selection': ['random', 'cyclic']
        }),

        ('decision_tree', Pipeline([
            ('scaler', StandardScaler()),  # scaler không ảnh hưởng đến cây nhưng giữ cho đồng nhất
            ('model', DecisionTreeRegressor())
        ]), {
            'model__criterion': ['squared_error', 'friedman_mse'],
            'model__splitter': ['best', 'random']
        })
    ]

    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    results = []

    for name, pipeline, params in models:
        gs = GridSearchCV(pipeline, params, cv=cv, return_train_score=False, n_jobs=-1, verbose=1)
        gs.fit(X, y)
        results.append({
            'model': name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

    return pd.DataFrame(results)

find_best_model_using_gridsearchcv(X, y)
```

```python
def predict_price(location, sqft, bath, bhk):
    loc_index = np.where(X.columns==location)[0][0]

    x = np.zeros(len(X.columns))
    x[0] = sqft
    x[1] = bath
    x[2] = bhk
    if loc_index >= 0:
        x[loc_index] = 1
    return lr_clf.predict(pd.DataFrame([x], columns=X.columns))[0] #cần [0] để ra số luôn chứ không thì sẽ ra array[(số đó)]
```
